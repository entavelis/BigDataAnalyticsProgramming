```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
rmse = pd.read_csv("RMSE.csv", header=None, sep=":")
rmse = rmse.drop(1,axis = 1)
rmse = rmse.drop(2, axis = 1)
rmse["thr"]= rmse[0].map(lambda x: float(x.split("adj")[0].split("thr")[1]))
rmse["adj"]= rmse[0].map(lambda x: int(x.split("adj")[1].split("lc")[0]))
rmse["lc"]= rmse[0].map(lambda x: int(x.split("lc")[1]))
rmse = rmse.drop(0, axis = 1)
# rmse = rmse.set_index(["thr","lc"]).drop(0,axis = 1)

# rmse.sort_values(3)

fig, ax = plt.subplots(figsize=(9,7))         # Sample figsize in inches
rmse = rmse[rmse["adj"] < 55].set_index(["thr","lc","adj"]).unstack()
rmse = rmse.xs(3, axis=1, drop_level=True)
sns.heatmap(rmse, annot=True, fmt=".3g",cbar=False)
plt.xlabel("Adjustment Threshold")
plt.ylabel("Threshold - Least Common")
# ax.xaxis.label.set_color('white')
# ax.yaxis.label.set_color('white')



plt.savefig("rmse.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
rmse = pd.read_csv("nnRMSE.csv", header=None, sep=":")
rmse = rmse.drop(1,axis = 1)
rmse = rmse.drop(2, axis = 1)
rmse["thr"]= rmse[0].map(lambda x: float(x.split("adj")[0].split("thr")[1]))
# rmse["adj"]= rmse[0].map(lambda x: int(x.split("adj")[1].split("lc")[0]))
rmse["lc"]= rmse[0].map(lambda x: int(x.split("lc")[1].split("k")[0]))
rmse["k"]= rmse[0].map(lambda x: int(x.split("k")[1]))
rmse = rmse.drop(0, axis = 1)
# rmse = rmse.set_index(["thr","lc"]).drop(0,axis = 1)

# rmse.sort_values(3)

rmse = rmse[rmse["lc"] < 56].set_index(["thr","k","lc"]).unstack()
rmse = rmse.xs(3, axis=1, drop_level=True)
fig, ax = plt.subplots(figsize=(9,7))         # Sample figsize in inches
sns.heatmap(rmse, annot=True, fmt=".3g",cbar=False)
plt.xlabel("Least Common Itemes")
plt.ylabel("Threshold - K of Neighbors")
fig.patch.set_facecolor('white')
plt.savefig("nnrmse.png")
plt.show()
```
