```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import colors
import pypandoc 

def background_gradient(s, m, M, cmap='PuBu', low=0, high=0):
    rng = M - m
    norm = colors.Normalize(m - (rng * low),
                            M + (rng * high))
    normed = norm(s.values)
    c = [colors.rgb2hex(x) for x in plt.cm.get_cmap(cmap)(normed)]
    return ['background-color: %s' % color for color in c]

def styler(df, maxis = None):
    s = df.style.apply(background_gradient,
               cmap='PuBu',
               m=df.min().min(),
               M=df.max().max(),
               low=0.5,
               high=1.0).highlight_max(axis= maxis).format("{:.3f}")
    return s

def weightedmean(df):
    weights = [10] + (df.index.values[1:] - df.index.values[:-1]).tolist()
    return df.apply(lambda b : b*weights).sum()/df.index.values[-1] 

def addf1(df):
    test = df
    test.columns = ["Acc","Prec","TNR","TPR"]
    test["F-score"] = 2*(test["Acc"]*test["Prec"])/(test["Acc"] + test["Prec"])   
    return test

def tolatex(s,outname="test"):
    print(s.render())
    output = pypandoc.convert_text(
        s.render(),
        'tex', format='html')
    print(output)
    
    with open(outname + ".tex",'w') as f:
        f.write(output)


idx = pd.IndexSlice
nbfh = pd.DataFrame()
nbcms = pd.DataFrame()
pfh = pd.DataFrame()
pcms = pd.DataFrame()
```

```python
def readata(extra="nbfh"):

    df = pd.DataFrame()
    extrapath = "/output/out" + extra + "/"
    for filename in os.listdir(os.getcwd() + extrapath):
        elements = filename.split("_")
        params = elements[-1].split(".")
        
        if (len(params) + len(elements) < 5):
            continue
            
        try:
            index = params[1:] + map(float,elements[1:-1]) + [float(params[0])]
        except:
            index = params[1:] +  map(float,elements[1:-1])
            
        mulindex = [[i] for i in index]
        
        fn = (os.getcwd() + extrapath + filename)
        
        if(index[1] == "corr" or index[1] == "acc-maj"): continue
        
        if (len(index) == 5  and index[0] == "nbcms"):
    #         print(mulindex[:-1])
            temp = pd.read_csv(fn,index_col=0,delimiter="\t",)
            if (len(temp)<273): continue
            temp.columns = pd.MultiIndex.from_arrays(mulindex[1:],names=("Metric","Max_N","Buckets","Hashes"))
#             nbcms = pd.concat([nbcms,temp],axis=1)
            df = pd.concat([df,temp],axis=1)
            
        elif (index[0] == "nbfh" and len(index) == 4):
    #         if (len(index) == 5):
    #             mulindex = mulindex[:-1]
                
    #         print(mulindex[:])
            temp = pd.read_csv(fn,index_col=0,delimiter="\t",)
            if (len(temp)<273): continue
            temp.columns = pd.MultiIndex.from_arrays(mulindex[1:],names=("Metric","Max_N","Buckets"))
            df = pd.concat([df,temp],axis=1)    
#             nbfh = pd.concat([nbfh,temp],axis=1)    
            
        elif (index[0] == "pfh") and len(index) == 6 and index[-1] == 2.0:
    #         if (len(index) == 5):
    #             continue
            mulindex = mulindex[:-1]
                
            temp = pd.read_csv(fn,index_col=0,delimiter="\t",)
            if (len(temp)<273): continue
            temp.columns = pd.MultiIndex.from_arrays(mulindex[1:],names=("Metric","Max_N","Buckets", "LearningRate"))
#             pfh = pd.concat([pfh,temp],axis=1)   
            df = pd.concat([df,temp],axis=1)   
        elif (index[0] == "pcms" and len(index) == 6):
    #         print(mulindex[:])
            temp = pd.read_csv(fn,index_col=0,delimiter="\t",)
            if (len(temp)<273): continue
            temp.columns = pd.MultiIndex.from_arrays(mulindex[1:],names=("Metric","Max_N","Buckets", "LearningRate", "Hashes"))
#             pcms = pd.concat([pcms,temp],axis=1)   
            df = pd.concat([df,temp],axis=1)   
    #     else:
    #         print(index)
    return df    
```

```python
def readnbfh(datecode = ""):

    df = pd.DataFrame()
    extrapath = "/output/outnbfh/"
    for filename in os.listdir(os.getcwd() + extrapath):
        elements = filename.split("_")
        
        if (elements[0] != "out" + datecode): continue
            
            
        params = elements[-1].split(".")
            
        if ((len(params) + len(elements)) < 4):
            continue
            
        try:
            index = params[2:] + map(float,elements[1:-1]) + [float(params[0]+ "." + params[1])]
        except:
            index = params[1:] +  map(float,elements[1:-1])
            
        
        fn = (os.getcwd() + extrapath + filename)
        
        if(index[1] == "corr" or index[1] == "acc-maj"): continue
        
            
#         if (len(index) == 4): index += [0.5]
        if (len(index) < 5): continue
            
        mulindex = [[i] for i in index]
        temp = pd.read_csv(fn,index_col=0,delimiter="\t",)
        if (len(temp)<273): continue
            
        temp.columns = pd.MultiIndex.from_arrays(mulindex[1:],names=("Metric","Max_N","Buckets","Threshold"))
        df = pd.concat([df,temp],axis=1)    
        
    return df    
```

```python
def readnbcms(datecode = ""):

    df = pd.DataFrame()
    extrapath = "/output/outnbcms/"
    for filename in os.listdir(os.getcwd() + extrapath):
        elements = filename.split("_")
        
        if (elements[0] != "out" + datecode): continue
            
            
        params = elements[-1].split(".")
            
        if ((len(params) + len(elements)) < 5):
            continue
            
        try:
            index = params[2:] + map(float,elements[1:-1]) + [float(params[0]+ "." + params[1])]
        except:
            index = params[1:] +  map(float,elements[1:-1])
            
        
        fn = (os.getcwd() + extrapath + filename)
        
        if(index[1] == "corr" or index[1] == "acc-maj"): continue
        
            
#         if (len(index) == 4): index += [0.5]
        if (len(index) < 6): continue
            
        mulindex = [[i] for i in index]
        temp = pd.read_csv(fn,index_col=0,delimiter="\t",)
        if (len(temp)<273): continue
            
        temp.columns = pd.MultiIndex.from_arrays(mulindex[1:],names=("Metric","Max_N","Buckets","Hashes","Threshold"))
        df = pd.concat([df,temp],axis=1)    
        
    return df    
```
